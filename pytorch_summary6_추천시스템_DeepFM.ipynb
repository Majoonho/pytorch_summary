### 3) Content-based Filtering과 Collaborative Filtering에 대해 설명하고 각각의 장단점에 대해 서술하시오.

* Content-based Filtering 은 유저가 구매했던, like 했던 또는 보았었던 상품을 대상을 기반으로 추천하는 것입니다.
    * 따라서 새로운 신제품에 대해서도 추천이 가능하고, 다른사람과의 독립된 요소로 영향을 받지 않습니다.
    * 개인의 독특한 취향도 반영할 수 있는 장점이 있습니다.
    * 반면 유저가 좋아하는 특성을 추출하는 방법이 복잡 할 수 있습니다.
*
*
* Collaborative Filtering 은 내가 구매했던 내역과 다른 사람이 구매했던 내역을 비교해서 상품을 추천하는 것입니다.
    * 비교로 추천하기 때문에 특징을 복잡하게 추출할 필요가 없다는 것이 장점입니다.
    * 반면 유저의 취향과 다른사람의 취향이 동일할 수 없으므로 정확한 추천이 안될 수 있습니다.
    * 유저의 독특한 취향을 반영하기 힘들며, 좀더 대중적이기 때문에 인기상품에 대해 추천할 가능성이 높아집니다.
